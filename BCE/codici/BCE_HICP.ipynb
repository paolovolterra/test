## [Inflazione e prezzi al consumo](https://www.ecb.europa.eu/stats/macroeconomic_and_sectoral/hicp/html/index.en.html) - HICP

```python
from ecbdata import ecbdata
import pandas as pd
import matplotlib.pyplot as plt

# https://data.ecb.europa.eu/data/datasets/ICP/ICP.M.U2.N.000000.4.ANR
# df = ecbdata.get_series('ICP.M.U2.N.000000.4.ANR', ) # 
df = ecbdata.get_series('ICP.M.IT.N.000000.4.ANR', start='2020-01') # Italia

df.TIME_PERIOD = pd.to_datetime(df.TIME_PERIOD)
df = df.set_index('TIME_PERIOD')
df.OBS_VALUE.plot()
plt.show()


'''Frequency                            Monthly [M]
Reference area                          Euro area (changing composition) [U2]
Adjustment indicator                    Neither seasonally nor working day adjusted [N]
Classification - ICP context            HICP - Overall index [000000]
Institution originating the data flow   Eurostat [4]
Series variation - ICP context          Annual rate of change [ANR]'''
```
