## 3a prova

- F2 = Currency and deposits of Non financial corpora...	
- F3 = totale debito (obbligazioni + prestiti)
- F4 = obbligazioni
- F7 = Financial derivatives of Non financial corpora..
- F3T4 su F = incidenza dei prestiti+obbligazioni sul totale attività/passività
- F8 = Other accounts payable of Non financial corpor..
- 

```python
import pandas as pd
import requests
from io import StringIO
```

```python
url = "https://data-api.ecb.europa.eu/service/data/QSA/Q.N.IT.W0.S11.S1.N.L.LE.F3.T._Z.XDC._T.S.V.N._T?format=csvdata"
res = requests.get(url)
res.raise_for_status()

lines = res.text.splitlines() # Identifica la riga header: cerca la prima riga che contiene "TIME_PERIOD"
header_row = None
for i, line in enumerate(lines):
    if "TIME_PERIOD" in line:
        header_row = i
        break

# Carica il CSV in un DataFrame saltando le righe prima dell’header
df = pd.read_csv(StringIO(res.text), skiprows=header_row)

df["PERIODO"] = df["TIME_PERIOD"].astype("period[Q]")
df.set_index("PERIODO", inplace=True)

df = df.rename(columns={"OBS_VALUE": "VALORE_MLN"})
df[['TITLE', 'REF_AREA',  'INSTR_ASSET', 'TIME_PERIOD', 'VALORE_MLN']]
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

df["VALORE_MLN"].plot(figsize=(12, 5), title="Obbligazioni emesse da Società Non Finanziarie (Italia)", grid=True)
plt.ylabel("Milioni di euro")
plt.tight_layout()
plt.show()
```

## 4a prova

```python
geo = 'DE'
```

```python
# F3 (Obbligazioni)
url_f3 = f"https://data-api.ecb.europa.eu/service/data/QSA/Q.N.{geo}.W0.S11.S1.N.L.LE.F3.T._Z.XDC._T.S.V.N._T?format=csvdata"
res_f3 = requests.get(url_f3) 
res_f3.raise_for_status()
lines_f3 = res_f3.text.splitlines()
header_f3 = next(i for i, line in enumerate(lines_f3) if "TIME_PERIOD" in line)
df_f3 = pd.read_csv(StringIO(res_f3.text), skiprows=header_f3)
df_f3["PERIODO"] = df_f3["TIME_PERIOD"].astype("period[Q]")
df_f3.set_index("PERIODO", inplace=True)
df_f3 = df_f3.rename(columns={"OBS_VALUE": "OBBLIGAZIONI_MLN"})
```

```python
# F4 (Prestiti)
url_f4 = f"https://data-api.ecb.europa.eu/service/data/QSA/Q.N.{geo}.W0.S11.S1.N.L.LE.F4.T._Z.XDC._T.S.V.N._T?format=csvdata"
res_f4 = requests.get(url_f4)
res_f4.raise_for_status()
lines_f4 = res_f4.text.splitlines()
header_f4 = next(i for i, line in enumerate(lines_f4) if "TIME_PERIOD" in line)
df_f4 = pd.read_csv(StringIO(res_f4.text), skiprows=header_f4)
df_f4["PERIODO"] = df_f4["TIME_PERIOD"].astype("period[Q]")
df_f4.set_index("PERIODO", inplace=True)
df_f4 = df_f4.rename(columns={"OBS_VALUE": "PRESTITI_MLN"})
```

```python
# 3. Unisci e calcola le percentuali
df_debito = df_f3[["OBBLIGAZIONI_MLN"]].join(df_f4[["PRESTITI_MLN"]], how="inner")
df_debito["TOTALE_MLN"] = df_debito["OBBLIGAZIONI_MLN"] + df_debito["PRESTITI_MLN"]
df_debito["%PRESTITI"] = (df_debito["PRESTITI_MLN"] / df_debito["TOTALE_MLN"]) * 100
df_debito["%OBBLIGAZIONI"] = (df_debito["OBBLIGAZIONI_MLN"] / df_debito["TOTALE_MLN"]) * 100
```

```python
# 4. Visualizza con grafico stacked
df_plot = df_debito[["%PRESTITI", "%OBBLIGAZIONI"]].copy()
df_plot.columns = ["Prestiti", "Obbligazioni"]

df_plot.plot(kind="bar", stacked=True, figsize=(14, 6))
plt.title("Composizione % dei debiti finanziari delle imprese")
plt.ylabel("Percentuale")
plt.xlabel("Periodo")
plt.legend(
    loc="upper center", bbox_to_anchor=(0.5, -0.15),
    ncol=2, frameon=False, fontsize="small"
)
plt.tight_layout()
plt.show()
```

```python
url_tot = f"https://data-api.ecb.europa.eu/service/data/QSA/Q.N.{geo}.W0.S11.S1.N.L.LE.F3T4.T._Z.XDC._T.S.V.N._T?format=csvdata"
res_tot = requests.get(url_tot)
res_tot.raise_for_status()
lines_tot = res_tot.text.splitlines()
header_tot = next(i for i, line in enumerate(lines_tot) if "TIME_PERIOD" in line)
df_tot = pd.read_csv(StringIO(res_tot.text), skiprows=header_tot)
df_tot["PERIODO"] = df_tot["TIME_PERIOD"].astype("period[Q]")
df_tot.set_index("PERIODO", inplace=True)
df_tot = df_tot.rename(columns={"OBS_VALUE": "F3T4_MLN"})
```

```python
df_debito = df_debito.join(df_tot[["F3T4_MLN"]], how="inner")
df_debito["DIFF_MLN"] = df_debito["TOTALE_MLN"] - df_debito["F3T4_MLN"]
```

```python
df_debito["%PRESTITI"] = (df_debito["PRESTITI_MLN"] / df_debito["F3T4_MLN"]) * 100
df_debito["%OBBLIGAZIONI"] = (df_debito["OBBLIGAZIONI_MLN"] / df_debito["F3T4_MLN"]) * 100
df_debito["%ALTRO"] = 100 - df_debito["%PRESTITI"] - df_debito["%OBBLIGAZIONI"]
```

```python
df_plot = df_debito[["%PRESTITI", "%OBBLIGAZIONI", "%ALTRO"]].copy()
df_plot.columns = ["Prestiti", "Obbligazioni", "Altro"]

df_plot.plot(kind="bar", stacked=True, figsize=(14, 6))
plt.title("Composizione % dei debiti finanziari delle imprese (fonte BCE - QSA)")
plt.ylabel("Percentuale sul totale debito (F3T4)")
plt.xlabel("Periodo")
plt.legend(
    loc="upper center", bbox_to_anchor=(0.5, -0.15),
    ncol=3, frameon=False, fontsize="small"
)
plt.tight_layout()
plt.show()
```
