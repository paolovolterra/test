Il messaggio flow_msg contiene:

```python
for k, v in sorted(flow_msg.dataflow.items(), key=lambda item: item[1].name.en.lower()):
    print(f"{k}: {v.structure.id} - {v.name.en}")
```

```python
for k in sorted(flow_msg.dataflow.keys()):
    print(f"{k}: {flow_msg.dataflow[k].structure.id} - {flow_msg.dataflow[k].name.en}")
```

```python
keywords = ["sector", "account", "financial", "non-financial", "debt"]

for k, v in flow_msg.dataflow.items():
    name = v.name.en.lower()
    if any(kw in name for kw in keywords):
        print(f"{k}: {v.structure.id} - {v.name.en}")
```

```python
import sdmx

# Crea il client ECB
ecb = sdmx.Client('ecb')

# Recupera il DataStructureDefinition per ECB_EXR1
dsd_msg = ecb.get('datastructure', resource_id='ECB_EXR1')
dsd = dsd_msg.structure['ECB_EXR1']

# Recupera le codelist disponibili nel messaggio
codelists = dsd_msg.codelist

# Per ogni dimensione, stampa i valori se associata a una codelist
for dim in dsd.dimensions:
    print(f"\nDimensione: {dim.id}")

    if dim.local_representation.enumerated:
        cl_id = dim.local_representation.enumerated.id
        print(f"  → Codelist: {cl_id}")

        cl = codelists[cl_id]
        for code in cl:
            print(f"    {code.id}: {code.name.en}")  # ← oppure una delle varianti sopra
    else:
        print("  → Nessuna codelist associata.")

```

```python
import sdmx
ecb = sdmx.Client('ecb')

# Scarica la struttura ESA (European System of Accounts)
dsd_msg = ecb.get('datastructure', resource_id='ECB_ESA1')
dsd = dsd_msg.structure['ECB_ESA1']
```

```python
# Estrai tutte le codelist
codelists = dsd_msg.codelist

# Visualizza dimensioni e codici
for dim in dsd.dimensions:
    print(f"Dimensione: {dim.id}")
    if dim.local_representation.enumerated:
        cl_id = dim.local_representation.enumerated.id
        cl = codelists[cl_id]
        for code in cl:
            print(f"  {code.id}: {code.name.en if hasattr(code.name, 'en') else ''}")
    else:
        print("  → Nessuna codelist associata.")
```

```python
import pandas as pd

key = '.USD+JPY...'
params = dict(startPeriod = '2016')
ecb = sdmx.Request('ECB')
data_msg = ecb.data('EXR', key=key, params = params)
data = data_msg.data[0]
daily = [s for sk, s in data.series.items() if sk.FREQ == 'D']
cur_df = pd.concat(sdmx.to_pandas(daily)).unstack().T
cur_df.index = pd.DatetimeIndex(cur_df.index,)
cur_df
```
