    Il **debito escluso il trade credit** (in inglese: *debt excluding trade credit*) √® una misura del **debito finanziario** delle imprese **che esclude i debiti commerciali**, ossia quelli verso fornitori e altri conti da pagare.

---

### üìå **Definizione tecnica**

Il debito delle societ√† **non finanziarie** (*S11*) **escludendo il trade credit** si basa sulla somma dei seguenti strumenti ESA:

| Codice | Voce                                               | Incluso? |
| ------ | -------------------------------------------------- | -------- |
| **F3** | Obbligazioni (*Debt securities*)                   | ‚úÖ S√¨     |
| **F4** | Prestiti (*Loans*)                                 | ‚úÖ S√¨     |
| **F6** | Schemi pensione/assicurativi (*Pension/insurance*) | ‚úÖ S√¨     |

üö´ **Escluso:**

* **F8** = *Altri conti da pagare* ‚Üí include **debiti commerciali** (*trade credit*) e imposte da pagare
* **F5** = Patrimonio netto (azioni/quote)
* **F7** = Derivati

---

### üìä Codifica nella BCE/Eurostat

* Nella **BCE** la voce **"Debt excluding trade credit"** √® spesso etichettata come:

  ```
  QSA.A.N.IT.W0.S11.S1.N.L.LE.FP.T._Z.XDC_R_B1GQ_CY._T.S.V.N._T
  ```

  dove `FP` = **F3 + F4 + F6**, e **esclude F8** (crediti commerciali).

---

### üîç Differenza tra:

| Tipo di Debito                       | Codice ESA  | Cosa include                           |
| ------------------------------------ | ----------- | -------------------------------------- |
| **Totale passivit√† finanziarie (F)** | F1‚ÄìF8       | Tutti gli strumenti finanziari         |
| **Debito finanziario (FPT)**         | F3+F4(+F6)  | Solo strumenti di debito veri e propri |
| **Debito incluso trade credit**      | F3+F4+F6+F8 | Include anche fornitori, imposte, ecc. |
| **Debito escluso trade credit**      | F3+F4+F6    | Esclude i debiti commerciali           |

---

### üß† Perch√© √® importante?

* **Pi√π adatto per confronti internazionali**: in molti paesi i debiti commerciali variano molto in funzione del sistema economico e delle abitudini di pagamento.
* **Pi√π fedele al rischio finanziario**: considera solo debiti con costo finanziario (prestiti, obbligazioni, ecc.).

---

### ‚úÖ Conclusione

**Debito escluso trade credit = F3 (obbligazioni) + F4 (prestiti) + F6 (pensioni/assicurazioni)**
üî∏ **Non considera F8** (crediti commerciali e altre passivit√† non finanziarie).
üî∏ √à usato dalla BCE, OCSE e FMI come indicatore "core" di indebitamento delle imprese.

### üîç Definizione ufficiale di **"debito finanziario" (FPT)** per le **societ√† non finanziarie (S11)**

Secondo BCE e Eurostat, la **definizione operativa di debito** nel contesto dei conti finanziari (dataset **QSA**) include:

| Codice ESA | Strumento finanziario                       | Incluso in FPT                                                    |
| ---------- | ------------------------------------------- | ----------------------------------------------------------------- |
| **F3**     | Obbligazioni                                | ‚úÖ S√¨                                                              |
| **F4**     | Prestiti                                    | ‚úÖ S√¨                                                              |
| **F6**     | Schemi pensionistici/assicurativi (es. TFR) | ‚úÖ S√¨ (in alcune definizioni ampie, ma opzionale secondo contesto) |



Il codice F (senza numero) aggrega **tutti gli strumenti finanziari** del sistema ESA:

| Codice | Voce                                           |
| ------ | ---------------------------------------------- |
| **F1** | Oro monetario e DSP (non rilevante per S11)    |
| **F2** | Moneta e depositi                              |
| **F3** | Obbligazioni                                   |
| **F4** | Prestiti                                       |
| **F5** | Azioni e quote di fondi di investimento        |
| **F6** | Schemi pensione/assicurazione                  |
| **F7** | Derivati finanziari e stock option             |
| **F8** | Altri conti da pagare (es. fornitori, imposte) |
