```python
import pandas as pd

# Carica e filtra i dataset GDP

# Filtra per titoli che contengono 'GDP' (case-sensitive)
filtro_gdp = df['title'].fillna('').str.contains('GDP', case=True, na=False)
df_gdp = df.loc[filtro_gdp].dropna(subset=['code'])

# Rimuovi duplicati di 'code' mantenendo solo il primo titolo associato
df_gdp = df_gdp.drop_duplicates(subset='code')

# Parsing delle date
df_gdp['last update of data'] = pd.to_datetime(df_gdp['last update of data'], format='%d.%m.%Y', errors='coerce')
df_gdp['data start'] = pd.to_datetime(df_gdp['data start'], errors='coerce')
df_gdp['data end'] = pd.to_datetime(df_gdp['data end'], errors='coerce')

# Calcola copertura e punteggio grezzo
df_gdp['copertura_anni'] = ((df_gdp['data end'] - df_gdp['data start']).dt.days / 365.25).fillna(0)
df_gdp['valori'] = df_gdp['values'].fillna(0)
df_gdp['update_recenti'] = (df_gdp['last update of data'] >= pd.Timestamp('2024-01-01')).astype(int)

# Punteggio: semplice somma pesata (modificabile)
df_gdp['score'] = (
    df_gdp['copertura_anni'].rank(ascending=False) * 0.4 +
    df_gdp['valori'].rank(ascending=False) * 0.4 +
    df_gdp['update_recenti'] * 20  # Boost se aggiornato di recente
)

# Top 10 per punteggio
df_top = df_gdp.sort_values(by='score', ascending=False).head(10)[['code', 'title', 'copertura_anni', 'valori', 'last update of data', 'score']]

# Tabella Markdown
from tabulate import tabulate
print(tabulate(df_top, headers='keys', tablefmt='github', showindex=False))
```

Il dataset **`teina110`** con titolo **"GDP deflator"** si riferisce a un indicatore economico fondamentale per l’analisi della crescita reale di un’economia:


### 🧮 **Cos'è il GDP deflator?**

Il **GDP deflator** (in italiano: *deflatore del PIL*) è un **indice dei prezzi implicito** che misura la variazione dei prezzi di tutti i beni e servizi prodotti all’interno di un’economia, cioè:

> **GDP deflator = (PIL nominale / PIL reale) × 100**


### 📊 Cosa rappresenta?

* **PIL nominale**: PIL calcolato a **prezzi correnti** (cioè ai prezzi dell’anno di riferimento)
* **PIL reale**: PIL calcolato a **prezzi costanti** (cioè rimuovendo l’effetto dell’inflazione)

Quindi il deflatore:

* riflette l’**andamento generale dei prezzi** nell’economia
* è un **indicatore di inflazione più ampio** rispetto all’IPCA o CPI, perché include tutti i beni e servizi del PIL (non solo quelli al consumo)
* **non ha una base fissa** (a differenza dell’indice CPI)


### ✅ Perché è importante?

* Usato per **deflazionare il PIL** e stimare la crescita economica reale
* Indicato spesso nei **documenti macroeconomici** (UPB, DEF, ISTAT)
* È un indicatore chiave per la **politica fiscale e monetaria**
