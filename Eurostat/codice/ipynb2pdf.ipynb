```python
import os
import subprocess
from pathlib import Path

# 📁 Cartella contenente i notebook (modifica se necessario)
cartella = Path("D:\PKM\Github\Eurostat\codice")  # cartella corrente

# 🔍 Cerca tutti i file .ipynb
notebooks = list(cartella.glob("*.ipynb"))

if not notebooks:
    print("❌ Nessun file .ipynb trovato nella cartella:", cartella.resolve())
else:
    print(f"🔎 Trovati {len(notebooks)} file .ipynb. Avvio conversione in PDF (webpdf)…")

    for nb in notebooks:
        output_pdf = nb.with_suffix(".pdf")
        print(f"📄 Convertendo: {nb.name} → {output_pdf.name}")

        try:
            subprocess.run([
                "jupyter", "nbconvert",
                str(nb),
                "--to", "webpdf",
                "--allow-chromium-download"
            ], check=True)
        except subprocess.CalledProcessError as e:
            print(f"⚠️ Errore nella conversione di {nb.name}: {e}")
        else:
            print(f"✅ {output_pdf.name} creato con successo.")

    print("🎉 Conversione completata.")
```
