```python
import os
import subprocess
from pathlib import Path

# ğŸ“ Cartella contenente i notebook (modifica se necessario)
cartella = Path("D:\PKM\Github\Eurostat\codice")  # cartella corrente

# ğŸ” Cerca tutti i file .ipynb
notebooks = list(cartella.glob("*.ipynb"))

if not notebooks:
    print("âŒ Nessun file .ipynb trovato nella cartella:", cartella.resolve())
else:
    print(f"ğŸ” Trovati {len(notebooks)} file .ipynb. Avvio conversione in PDF (webpdf)â€¦")

    for nb in notebooks:
        output_pdf = nb.with_suffix(".pdf")
        print(f"ğŸ“„ Convertendo: {nb.name} â†’ {output_pdf.name}")

        try:
            subprocess.run([
                "jupyter", "nbconvert",
                str(nb),
                "--to", "webpdf",
                "--allow-chromium-download"
            ], check=True)
        except subprocess.CalledProcessError as e:
            print(f"âš ï¸ Errore nella conversione di {nb.name}: {e}")
        else:
            print(f"âœ… {output_pdf.name} creato con successo.")

    print("ğŸ‰ Conversione completata.")
```
