## Grafico 1 - PIL reale (volume) | namq_10_gdp

- PIL reale da fine 1999: la sovraperformance dell’economia spagnola non è una novità
- Nota: 1999 4° trim.= 100; Fonte: Intesa Sanpaolo, Eurostat
- https://ec.europa.eu/eurostat/api/dissemination/statistics/1.0/data/namq_10_gdp?geo=IT&s_adj=SCA&unit=CLV_MEUR&na_item=B1GQ

![](./media/20250707112807.png)

```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append("D:/PKM/Github/Eurostat/codice")  # Aggiunge il path dove si trova il modulo
from eurostat_utils import fetch_eurostat_data # questa è la mia funzione per estrarre i json
df = fetch_eurostat_data("namq_10_gdp", {
    'na_item': 'B1GQ',
    'unit': 'CLV20_MEUR',
    's_adj': 'SCA',
    'geo': ['IT', 'ES', 'FR', 'DE'],
    'na_item': 'B1GQ'
})
# Filtra per trimestre
df['time'] = pd.to_datetime(df['time'], format='%Y-Q%q')
df = df[df['time'] >= '1999-10-01']  # 1999Q4
# Pivot: righe = tempo, colonne = paese
pivot_df = df.pivot(index='time', columns='geo_label', values='value').dropna()
# Normalizza base 100 a 1999Q4
base = pivot_df.iloc[0]
norm_df = pivot_df.divide(base).multiply(100)

labels = ['Italy', 'Germany', 'France', 'Spain']
colors = ['navy', 'darkorange', 'seagreen', 'gold']

plt.figure(figsize=(12, 5))
for label, color in zip(labels, colors):
    plt.plot(norm_df.index, norm_df[label], label=label, linewidth=4.5, color=color)

plt.ylabel("Index (1999 Q4 = 100)")
plt.title("PIL reale (volume) – Italia, Germania, Francia, Spagna")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Grafico 2 - PIL reale indicizzato | namq_10_gdp

Dopo un avvio lento, la ripresa post-pandemica in Spagna è stata superiore a quella registrata in Italia (e nelle altre principali economie europee)

Il grafico mostra il PIL reale indicizzato al 4° trimestre 2019 = 100, con i dati trimestrali per:

- Italia (IT)
- Germania (DE)
- Francia (FR)
- Spagna (ES)

e copre il periodo dal 2019Q4 fino al 2025.

![](./media/Pasted%20image%2020250707122520.png)

```python
import pandas as pd
import matplotlib.pyplot as plt
from eurostat_utils import fetch_eurostat_data

# Estrai dati trimestrali dal 2019 in poi
df = fetch_eurostat_data("namq_10_gdp", {
    'na_item': 'B1GQ',
    'unit': 'CLV20_MEUR',
    's_adj': 'SCA',
    'geo': ['IT', 'DE', 'FR', 'ES']
})

# Converti il campo tempo
df = df[df['time'] >= '2019-10-01']  # dal Q4 2019 in poi

# Crea tabella pivot: righe = tempo, colonne = paesi
pivot_df = df.pivot(index='time', columns='geo_code', values='value')

# Indicizza a 2019Q4 = 100
base = pivot_df.loc[pd.Timestamp('2019-10-01')]
norm_df = pivot_df.divide(base).multiply(100)

# Mappa etichette e colori
labels = {'IT': 'Italy', 'DE': 'Germany', 'FR': 'France', 'ES': 'Spain'}
colors = {'IT': 'navy', 'DE': 'darkorange', 'FR': 'seagreen', 'ES': 'gold'}

# Plot
plt.figure(figsize=(10, 4))
for code in ['IT', 'DE', 'FR', 'ES']:
    plt.plot(norm_df.index, norm_df[code], label=labels[code], linewidth=3.5, color=colors[code])

# Valori finali (ultimo punto)
for code in ['IT', 'DE', 'FR', 'ES']:
    val = norm_df[code].iloc[-1]
    plt.text(norm_df.index[-1], val, f"{val:.1f}", fontsize=12, color=colors[code], va='center', ha='left')

plt.ylim(75, 110)
plt.ylabel("Index (2019 Q4 = 100)")
plt.title("PIL reale indicizzato – Italia, Germania, Francia, Spagna")
plt.grid(True)
plt.legend(loc='lower left')
plt.tight_layout()
plt.show()
```

## Grafico 3 - PIL reale pro capite | namq_10_pc

- Gross domestic product (GDP) and main components per capita
- PIL pro-capite, 2019 4° trim. = 100. Fonte: Intesa Sanpaolo, Eurostat
- https://ec.europa.eu/eurostat/api/dissemination/statistics/1.0/data/namq_10_pc?geo=IT&unit=CLV20_EUR_HAB&na_item=B1GQ
- dimensioni: "freq","unit","s_adj","na_item","geo","time"

![](./media/Pasted%20image%2020250707125258.png)


 Il dataset namq_10_pc è l'opzione ufficiale e diretta di Eurostat per il PIL pro capite, espresso in catene di volume (reale). Ti permette di replicare con precisione il grafico PIL reale pro capite, 2019Q4 = 100, senza calcoli manuali.

Contiene il PIL reale pro capite (quindi già diviso per abitanti) in diverse unità:

- CLV20_EUR_HAB → Chain-linked volumes (base 2020), euro per abitante (reale)
- CP_EUR_HAB → Corrente, euro per abitante (nominale)
- CLV_PCH_SM → Variazione % reale anno su anno per abitante

#### unit

- "CP_EUR_HAB": "Current prices, euro per capita",
- "CP_NAC_HAB": "Current prices, units of national currency per capita",
- "CLV_I20_HAB": "Chain linked volumes, Index 2020=100, per capita",
- "CLV_I15_HAB": "Chain linked volumes, Index 2015=100, per capita",
- "CLV_I10_HAB": "Chain linked volumes, Index 2010=100, per capita",
- "**CLV20_EUR_HAB**": "Chain linked volumes (2020), euro per capita",
- "CLV15_EUR_HAB": "Chain linked volumes (2015), euro per capita",
- "CLV10_EUR_HAB": "Chain linked volumes (2010), euro per capita",
- "CLV_PCH_SM_HAB": "Chain linked volumes, percentage change compared to same period in previous year, p- capita"


#### na_item

- "**B1GQ**": "Gross domestic product at market prices",
- "P3": "Final consumption expenditure",
- "P3_S13": "Final consumption expenditure of general government",
- "P31_S13": "Individual consumption expenditure of general government",
- "P32_S13": "Collective consumption expenditure of general government",
- "P31_S14_S15": "Household and NPISH final consumption expenditure",
- "P31_S14": "Final consumption expenditure of households",
- "P31_S15": "Final consumption expenditure of NPISH",
- "P41": "Actual individual consumption"

```python
import pandas as pd
import matplotlib.pyplot as plt
from eurostat_utils import fetch_eurostat_data

# Estrai PIL pro capite reale trimestrale, Italia e Spagna
# https://ec.europa.eu/eurostat/api/dissemination/statistics/1.0/data/namq_10_pc?geo=IT&unit=CLV20_EUR_HAB&na_item=B1GQ
df = fetch_eurostat_data("namq_10_pc", {
    'na_item': 'B1GQ',
    'unit': 'CLV20_EUR_HAB',  # PIL reale pro capite (chain-linked volume)
    's_adj': 'NSA',
    'geo': ['IT', 'ES']
})

# Filtra dal Q4 2019 in poi
df = df[df['time'] >= '2019-10-01']

# Pivot: righe = tempo, colonne = paesi
pivot_df = df.pivot(index='time', columns='geo_code', values='value')

# Indicizza a base 100 = 2019Q4
base = pivot_df.loc[pd.Timestamp('2019-10-01')]
norm_df = pivot_df.divide(base).multiply(100)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(norm_df.index, norm_df['IT'], label='IT', color='navy', linewidth=3)
plt.plot(norm_df.index, norm_df['ES'], label='ES', color='darkorange', linewidth=3)

# Etichette finali
for code, color in [('IT', 'navy'), ('ES', 'darkorange')]:
    val = norm_df[code].iloc[-1]
    plt.text(norm_df.index[-1], val, f'{val:.1f}', color=color, fontsize=14, weight='bold')

plt.ylim(75, 110)
plt.ylabel("Index (2019 Q4 = 100)")
plt.title("PIL reale pro capite – Italia vs Spagna")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

- Labour productivity and unit labour costs

- namq_10_lp_ulc

![](media/Fig04.png)



```python
import pandas as pd
import matplotlib.pyplot as plt
from eurostat_utils import fetch_eurostat_data
# Estrai produttività reale per ora lavorata
df = fetch_eurostat_data("namq_10_lp_ulc", {
    'na_item': 'RLPR_HW',
    'unit': 'I10',
    'geo': ['DE', 'FR', 'ES', 'IT'],
    's_adj': 'SCA' # dati destagionalizzati e corretti per effetti di calendario (SCA)
})
# 2. Parsing corretto della variabile temporale
df['time'] = pd.to_datetime(df['time_label'], errors='coerce')
df.loc[df['time'].isna() & df['time_label'].str.contains('Q'), 'time'] = pd.PeriodIndex(
    df.loc[df['time'].isna(), 'time_label'], freq='Q').to_timestamp()
# 3. Filtro dal 1999 Q4 in poi
df = df[df['time'] >= '1999-10-01']
# Pivot: tempo = index, paesi = colonne
pivot_df = df.pivot(index='time', columns='geo_code', values='value')
# Indicizza a 1999Q4 = 100
base = pivot_df.loc[pd.Timestamp('1999-10-01')]
norm_df = pivot_df.divide(base).multiply(100)

# Etichette e colori
labels = {'DE': 'Germany', 'ES': 'Spain', 'FR': 'France', 'IT': 'Italy'}
colors = {'DE': 'navy', 'ES': 'darkorange', 'FR': 'seagreen', 'IT': 'gold'}
# Plot
plt.figure(figsize=(10, 5))
for code in ['DE', 'ES', 'FR', 'IT']:
    plt.plot(norm_df.index, norm_df[code], label=labels[code], linewidth=7, color=colors[code])

plt.ylabel("Index (1999 Q4 = 100)")
plt.title("Produttività reale per ora lavorata – Germania, Francia, Spagna, Italia")
plt.grid(True, alpha=0.1)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from eurostat_utils import fetch_eurostat_data

# Estrai produttività reale per ora lavorata (indice 2010=100)
df = fetch_eurostat_data("namq_10_lp_ulc", {
    'na_item': 'RLPR_HW',
    'unit': 'I10',
    'geo': ['DE', 'FR', 'ES', 'IT'],
    's_adj': 'SCA'
})

# Parsing corretta della variabile temporale
df['time'] = pd.to_datetime(df['time_label'], errors='coerce')
df.loc[df['time'].isna() & df['time_label'].str.contains('Q'), 'time'] = pd.PeriodIndex(
    df.loc[df['time'].isna(), 'time_label'], freq='Q').to_timestamp()

# Filtra dal Q4 1999 in poi
df = df[df['time'] >= '1999-10-01']

# Pivot: tempo = index, paesi = colonne
pivot_df = df.pivot(index='time', columns='geo_code', values='value')

# Indicizza base 1999 Q4 = 100
base = pivot_df.loc[pd.Timestamp('1999-10-01')]
norm_df = pivot_df.divide(base).multiply(100)

# Etichette leggibili e colori simili a Intesa Sanpaolo
labels = {'DE': 'Germany', 'FR': 'France', 'IT': 'Italy', 'ES': 'Spain'}
colors = {'DE': 'navy', 'FR': 'seagreen', 'IT': 'gold', 'ES': 'darkorange'}

# Plot
plt.figure(figsize=(10, 6))
for code in ['DE', 'FR', 'IT', 'ES']:
    plt.plot(norm_df.index, norm_df[code], label=labels[code], linewidth=6, color=colors[code])

# Titolo e asse
plt.title("Produttività reale per ora lavorata", fontsize=13)
plt.ylabel("Indice (1999 Q4 = 100)")
plt.ylim(95, 130)
plt.grid(True, axis='y', alpha=0.2)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
# plt.legend(frameon=False)
# plt.legend(loc='upper left', bbox_to_anchor=(1, 1), frameon=False)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=4, frameon=False, fontsize=20)
plt.tight_layout()
plt.show()
```

## Grafico 5 - Inflazione | prc_hicp_manr

- prc_hicp_manr
- https://ec.europa.eu/eurostat/api/dissemination/statistics/1.0/data/prc_hicp_manr?time=2023-01
- dimensioni: ["freq","unit","coicop","geo","time"]
- frequenza: mensile
- esempio: _Foresti, G. (2024, giugno 18). Lo scenario per le imprese italiane: Focus Marche_


Il dataset **`prc_hicp_manr`** (HICP - **monthly annual rates of change**, cioè **variazioni tendenziali mensili**) serve *solo* per rappresentare **l'inflazione annuale mese per mese**, ovvero:

> **Variazione % rispetto allo stesso mese dell’anno precedente**, per ciascun mese.

È utile per:

* confronti temporali coerenti (es. inflazione di aprile 2025 vs aprile 2024),
* confronti internazionali (i dati sono armonizzati),
* analisi stagionali (es. inflazione di Natale o estiva),
* confronti tra aggregati COICOP (alimentari, energia, core, ecc.).

**Non va usato** per:

* ricostruire un indice dei prezzi (serve `prc_hicp_midx`),
* calcolare inflazione cumulata (serve l’indice base),
* analisi di contributi alla crescita (servono dati con pesi e indici puntuali).

Per rappresentare l'inflazione core usare uno di questi due approcci, entrambi ufficiali:

✅ Opzione 1: TOT_X_NRG_FOOD
Codice: "TOT_X_NRG_FOOD"

Descrizione: Overall index excluding energy and food, alcohol and tobacco

✅ Opzione 2: TOT_X_NRG_FOOD_NP
Codice: "TOT_X_NRG_FOOD_NP"

Descrizione: Overall index excluding energy and unprocessed food

È quella usata spesso da Intesa Sanpaolo e nei grafici macro italiani (come quello che stai replicando).



### 🔹 **HICP**

**(Harmonised Index of Consumer Prices)**
È l’indice armonizzato dei prezzi al consumo:

* Misura l’**inflazione generale**.
* Include **tutti i beni e servizi**, compresi **energia** e **generi alimentari non trasformati**.



### 🔸 **CORE**

**(HICP excl. energy & unprocessed food)**
È una misura di **inflazione di fondo** (core inflation):

* Esclude le componenti **più volatili**, come:

  * Energia (elettricità, gas, carburanti)
  * Cibo non lavorato (frutta, verdura fresche, carne, ecc.)
* È usata spesso per valutare l’**inflazione strutturale**, quella su cui la **BCE può agire meglio** con la politica monetaria.



> Questo confronto è importante perché:
>
> * La **HICP** può calare rapidamente con i prezzi dell’energia.
> * La **CORE** è più lenta a reagire e indica la **persistenza** delle pressioni inflazionistiche.

```python
import pandas as pd
import matplotlib.pyplot as plt
from eurostat_utils import fetch_eurostat_data

# === Estrai solo HICP headline (indice armonizzato dei prezzi al consumo) ===
df_hicp = fetch_eurostat_data("prc_hicp_manr", {
    'geo': 'EA',
    'coicop': 'CP00',
    'unit': 'RCH_A'
})

# Parsing colonna temporale
df_hicp['time'] = pd.to_datetime(df_hicp['time_label'], errors='coerce')

# Filtro dal 2019 in poi
df_hicp = df_hicp[df_hicp['time'] >= '2019-01-01']

# === Plot ===
plt.figure(figsize=(9, 7))
plt.plot(df_hicp['time'], df_hicp['value'], label='HICP', color='navy', linewidth=3)
plt.axhline(2, color='black', linestyle='--', linewidth=1)
plt.axhline(0, color='black', linewidth=1)
plt.ylim(-2, 11)
plt.ylabel("y/y % change", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='lower center', fontsize=20, frameon=False)
plt.title("Area euro: l’inflazione potrebbe scendere sotto il 2% nel 2025", fontsize=16)
plt.grid(True, alpha=0.1)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from eurostat_utils import fetch_eurostat_data

# === Estrai HICP headline (tutti i beni) ===
df_hicp = fetch_eurostat_data("prc_hicp_manr", {
    'geo': 'IT',
    'coicop': 'CP00',
    'unit': 'RCH_A'
})

# === Estrai HICP core: esclusi energia e alimentari non lavorati ===
df_core = fetch_eurostat_data("prc_hicp_manr", {
    'geo': 'EA',
    'coicop': 'TOT_X_NRG_FOOD_NP', # TOT_X_NRG_FOOD_NP, TOT_X_NRG_FOOD
    'unit': 'RCH_A'
})

# Parsing data
df_hicp['time'] = pd.to_datetime(df_hicp['time_label'])
df_core['time'] = pd.to_datetime(df_core['time_label'])

# Merge
df = pd.merge(
    df_hicp[['time', 'value']].rename(columns={'value': 'HICP'}),
    df_core[['time', 'value']].rename(columns={'value': 'CORE'}),
    on='time'
)

# Filtro dal 2019
df = df[df['time'] >= '2019-01-01']

# Plot
plt.figure(figsize=(9, 5))
plt.plot(df['time'], df['HICP'], label='HICP', color='navy', linewidth=3)
plt.plot(df['time'], df['CORE'], label='HICP CORE (ex energy & unprocessed food)', color='darkorange', linewidth=3)
plt.axhline(2, color='black', linestyle='--', linewidth=1)
plt.axhline(0, color='black', linewidth=1)
plt.ylim(-2, 14)
plt.ylabel("y/y % change", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.legend(loc='lower center', fontsize=20, frameon=False)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2, fontsize=14, frameon=False)

plt.title("Italia: inflazione", fontsize=16)
plt.grid(True, alpha=0.1)
plt.tight_layout()
plt.show()
```

```python
# Ricarico librerie e funzione personalizzata dopo reset ambiente
import pandas as pd
import matplotlib.pyplot as plt
from eurostat_utils import fetch_eurostat_data

# Voci COICOP da confrontare
coicop_codes = {
    "CP01": "Food & non-alcoholic beverages",
    "CP04": "Housing, water, electricity, gas",
    "CP07": "Transport",
    "CP11": "Restaurants & hotels"
}

# Caricamento dati per ciascuna voce
dfs = []
for code, label in coicop_codes.items():
    df = fetch_eurostat_data("prc_hicp_manr", {
        'geo': 'IT',
        'coicop': code,
        'unit': 'RCH_A'
    })
    df['label'] = label
    dfs.append(df[['time', 'value', 'label']])

# Unione dei dati
df_all = pd.concat(dfs)
df_all = df_all[df_all['time'] >= '2019-01-01']

# Grafico
plt.figure(figsize=(10, 6))
for label in df_all['label'].unique():
    subset = df_all[df_all['label'] == label]
    plt.plot(subset['time'], subset['value'], label=label, linewidth=2)

plt.title("Italia: inflazione YoY per settori selezionati", fontsize=16)
plt.axhline(2, color='gray', linestyle='dashed', linewidth=1)
plt.xlabel("")
plt.ylabel("y/y % change", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.25), ncol=2, fontsize=14, frameon=False)
plt.grid(True, axis='y', linestyle=':', alpha=0.5)
plt.tight_layout()
plt.show()
```

## Grafico 6 - investimenti | nama_10_gdp


- nama_10_gdp
- Gross domestic product (GDP) and main components (output, expenditure and income)
- esempio: _Foresti, G. (2024, giugno 18). Lo scenario per le imprese italiane: Focus Marche._
- https://ec.europa.eu/eurostat/api/dissemination/statistics/1.0/data/nama_10_gdp?geo=IT&time=2023

![](./media/investimenti.png)

_(2016=100; valori concatenati)_

Usiamo il dataset **`nama_10_gdp`** di Eurostat per analizzare gli **investimenti fissi lordi** perché:

### **1. Contiene tutte le componenti principali del PIL**

Il dataset `nama_10_gdp` include, oltre al PIL (`B1GQ`), anche tutte le **principali componenti della domanda aggregata**, tra cui:

* **`P51G`** = **Gross Fixed Capital Formation (GFCF)** = **Investimenti fissi lordi**
* Consumi (`P3`)
* Esportazioni (`P6`)
* Importazioni (`P7`)
* Variazione scorte (`P52 + P53 - P6M`)
* Redditi e produzione (output, compensation, etc.)


### **2. Disponibile in formato concatenato (chain-linked volumes)**

Per analisi in **valori reali (volumi)**, Eurostat fornisce:

* **`unit=CLV_I10`**: volumi concatenati, **indice base 2010 (che possiamo ricalibrare al 2016=100)**
* oppure **`CLV10_MEUR`**: volumi concatenati in milioni di euro a prezzi 2010

Questo è ciò che ci permette di confrontare gli investimenti tra Paesi **al netto dell’inflazione**.


### **3. Copertura temporale completa e dati armonizzati**

`nama_10_gdp` è la fonte ufficiale per confronti tra Paesi europei, con:

* dati armonizzati secondo ESA2010,
* serie lunghe dal 1995,
* aggiornamento trimestrale e annuale.



### Perché **non** usare dataset alternativi?

Altri dataset come:

* `nama_10_an6`: troppo disaggregati per macro-confronti,
* `nasq_10_nf_tr`: settoriali, più adatti per analisi settoriali,
* `nama_10_pc`: contiene solo **tassi di crescita**, non gli indici concatenati.



### In sintesi:

Usiamo **`nama_10_gdp` + `na_item=P51G` + `unit=CLV_I10`**
perché ci dà esattamente:

> **Gli investimenti fissi lordi reali, concatenati, comparabili nel tempo e tra Paesi.**


```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append("D:/PKM/Github/Eurostat/codice")  # Aggiunge il path dove si trova il modulo
from eurostat_utils import fetch_eurostat_data # questa è la mia funzione per estrarre i json
df = fetch_eurostat_data("nama_10_gdp", {
    "na_item": "P51G",       # Gross fixed capital formation
    "unit": "CLV_I10",       # Chain linked volumes, index 2010
    "geo": ["IT", "DE", "ES", "FR"],
})

# 2. Aggiungi colonna 'year'
df["year"] = pd.to_datetime(df["time"]).dt.year

# 3. Calcola l'indice con base 2016 = 100
df_base = df[df["year"] == 2016][["geo_code", "value"]].rename(columns={"value": "base_2016"})
df = df.merge(df_base, on="geo_code")
df["value_base2016"] = df["value"] / df["base_2016"] * 100

# 4. Filtra solo gli anni dal 2016 in poi
df = df[df["year"] >= 2016]

# 5. Prepara i dati per il grafico
df_pivot = df.pivot(index="year", columns="geo_label", values="value_base2016")
df_pivot = df_pivot[["Italy", "Germany", "Spain", "France"]]
df_pivot.columns = ["Italia", "Germania", "Spagna", "Francia"]
```

```python
anno_max = df['year'].max()
df[df['year'] == anno_max][['na_item_label','geo_label','year','value','base_2016']]
```

```python
# 6. Crea il grafico
plt.figure(figsize=(10, 6))
colori = {
    "Italia": "darkorange",
    "Germania": "navy",
    "Spagna": "seagreen",
    "Francia": "gold"
}

for paese in df_pivot.columns:
    plt.plot(df_pivot.index, df_pivot[paese], label=paese, color=colori[paese], linewidth=5.5)
    plt.text(
        x=df_pivot.index.max() + 0.1,
        y=df_pivot.loc[df_pivot.index.max(), paese],
        s=f"{df_pivot.loc[df_pivot.index.max(), paese]:.1f}",
        color=colori[paese],
        va="center",
        fontweight="bold"
    )

plt.title("Investimenti fissi lordi (2016=100)", fontsize=14)
plt.ylabel("Indice (2016=100)")
plt.xlabel("")
plt.xticks(df_pivot.index)
plt.ylim(90, 150)
plt.grid(axis='y', linestyle='--', alpha=1)
plt.legend(title="", loc="upper left")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import sys

sys.path.append("D:/PKM/Github/Eurostat/codice") # Percorso al tuo modulo locale
from eurostat_utils import fetch_eurostat_data 
df = fetch_eurostat_data("nama_10r_2gfcf", {
    "currency": "MIO_EUR",
    "sector": "S1",
    "nace_r2": "TOTAL",
    "geo": ["ITF1","ITF2","ITF3","ITF4","ITF5","ITF6","ITG1","ITG2"]
})
df["year"] = pd.to_datetime(df["time"]).dt.year # 2. Estrai l'anno
df_base = df[df["year"] == 2016][["geo_code", "value"]].rename(columns={"value": "base_2016"})
df = df.merge(df_base, on="geo_code")
df["value_base2016"] = df["value"] / df["base_2016"] * 100
df = df[df["year"] >= 2016]
```

```python
# 3. Pivot per grafico con indice base 2016 = 100
df_pivot = df.pivot(index="year", columns="geo_label", values="value_base2016")

# 4. Grafico matplotlib
plt.figure(figsize=(12, 10))
for regione in df_pivot.columns:
    plt.plot(df_pivot.index, df_pivot[regione], label=regione, linewidth=3.5)

plt.title("Investimenti fissi lordi (2016 = 100) – Regioni del Sud (NUTS 2)", fontsize=14)
plt.ylabel("Indice (base 2016 = 100)")
plt.xlabel("Anno")
plt.xticks(df_pivot.index)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title="", loc="upper left", fontsize=9)
plt.tight_layout()
plt.show()
```
