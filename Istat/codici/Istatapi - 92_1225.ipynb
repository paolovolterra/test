## 22_289|Resident Population over time


```python
from nbdev import *
from istatapi import *
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# set matplotlib themes
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = [16, 5]
#fiveThirtyEight palette
colors = ['#30a2da', '#fc4f30', '#e5ae38', '#6d904f', '#8b8b8b']
```

```python
all_available_df = discovery.all_available()
all_available_df.query('df_id == "22_289"')
```

```python
# initialize the dataset and get its dimensions
ds = DataSet(dataflow_identifier="22_289")
```

```python
ds.dimensions_info()
```

```python
dimension = "SESSO" #use "dimension" column from above
ds.get_dimension_values(dimension)
```

```python
freq = "A" #Annual frequency
itter107 = 'IT' #Italy
stacivx = "99" #total marital status
sesso = "9" #total

ds.set_filters(freq = freq, itter107 = itter107, stacivx = stacivx, sesso = sesso)
```

```python
residents_df = get_data(ds)
residents_df.head()
```

```python
import numpy as np
def tweak_residents_df(residents_df):
    residents_df = residents_df.query('CLASSE_ETA != "TOTAL"').copy()
    
    residents_df = residents_df.rename(columns={"TIME_PERIOD": "year"})
    residents_df['year'] = pd.to_datetime(residents_df['year'], format='%Y')
    residents_df['CLASSE_ETA'] = residents_df['CLASSE_ETA'].astype(str).str.replace('Y', '').str.replace('_GE100', '100').astype(int)
    residents_df['ETA_bin'] = pd.cut(residents_df['CLASSE_ETA'], bins= [0, 25, 45, 65, np.inf], labels=['0-25', '25-45', '45-65', '65+'])
    return (residents_df
            .groupby(['year', 'ETA_bin']).sum().reset_index()
            )

residents_df2 = tweak_residents_df(residents_df)
```

```python
# Plot the data
sns.lineplot(data=residents_df2, x="year", y="OBS_VALUE", hue="ETA_bin", marker = 'X')
plt.rcParams["figure.figsize"] = (15, 4)
plt.ticklabel_format(style='plain', axis='y')
```

```python
# initialize the dataset and get its dimensions
ds = DataSet(dataflow_identifier="22_289")

# 22_289
# 119_367
```

```python
ds.dimensions_info()
```

```python
dimension = "TIPO_INDDEM"
ds.get_dimension_values(dimension)
```

```python
ateco_2007 = ["0014", "J"] #TOTAL AND ITC services 
tipo_dato = "SERV_TURN_2" #type of data
adjustment = "Y" #seasonally adjusted data

ds.set_filters(ateco_2007=ateco_2007, tipo_dato=tipo_dato, adjustment=adjustment)
```

```python
service_df = get_data(ds)
service_df.head()
```

```python
# Plot the data
is_0014 = service_df["ATECO_2007"] == "0014"
is_J = service_df["ATECO_2007"] == "J"
after_2013 = service_df["TIME_PERIOD"] >= "2013"


plt.plot(
    "TIME_PERIOD",
    "OBS_VALUE",
    data=service_df[is_0014 & after_2013],
    marker="",
    color="blue",
    linewidth=2,
)
plt.plot(
    "TIME_PERIOD",
    "OBS_VALUE",
    data=service_df[is_J & after_2013],
    marker="",
    color="red",
    linewidth=1,
)
plt.axhline(y=100, color="black", linewidth=0.5, linestyle="-")
plt.legend()
plt.rcParams["figure.figsize"] = (15, 4)
```
