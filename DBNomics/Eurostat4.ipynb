## [ A.CLV05 _ ​MEUR.B1G.IT ] Annuale – Volumi concatenati (2005), milioni di euro – Valore aggiunto lordo – Italia
https://db.nomics.world/Eurostat/NAMA_10_GDP/A.CLV05_MEUR.B1G.IT?tab=chart

```python

import pandas as pd
df = pd.read_csv('https://api.db.nomics.world/v22/series/Eurostat/NAMA_10_GDP/A.CLV05_MEUR.B1G.IT?format=csv&observations=1')
df.columns
```

```python
df.columns = ['period', 'value_added_gross_million_euro']
df
```

```python
import matplotlib.pyplot as plt

# Assicurati che la colonna 'period' sia in formato datetime, se necessario
df['period'] = pd.to_datetime(df['period'], errors='coerce')

# Ordina per periodo (opzionale ma consigliato)
df = df.sort_values('period')

# Crea il grafico
plt.figure(figsize=(10, 5))
plt.plot(df['period'], df['value_added_gross_million_euro'], marker='o', linewidth=2)

plt.title('Valore aggiunto lordo in euro chain-linked (2005)')
plt.xlabel('Periodo')
plt.ylabel('Milioni di euro (prezzi 2005)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='period', y='value_added_gross_million_euro', marker='o')

plt.title('Valore aggiunto lordo in euro chain-linked (2005)')
plt.xlabel('Periodo')
plt.ylabel('Milioni di euro (prezzi 2005)')
plt.tight_layout()
plt.show()
```
