```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def plot_blocco_nuts(loc_ctp_list, titolo_blocco=None):
    # Filtro
    filtered = data[
        (data['LOC_CTP'].isin(loc_ctp_list)) &
        (data['ENTE_SEGN'] == '1070001') &
        (data['DESINV'].isin([905])) &
        (data['TIPTASSO'].isin([80, 800]))
    ][['DATA_OSS', 'LOC_CTP', 'TIPTASSO', 'VALORE']].copy()

    # Data come datetime
    filtered['DATA_OSS'] = pd.to_datetime(filtered['DATA_OSS'])

    # Normalizzazione base 100 per LOC_CTP + TIPTASSO
    filtered['VALORE_NORM'] = (
        filtered.sort_values('DATA_OSS')
        .groupby(['LOC_CTP', 'TIPTASSO'])['VALORE']
        .transform(lambda x: (x / x.iloc[0]) * 100)
    )

    # Reshape per plot
    melted = filtered.melt(
        id_vars=['DATA_OSS', 'LOC_CTP', 'TIPTASSO'],
        value_vars=['VALORE', 'VALORE_NORM'],
        var_name='Tipo',
        value_name='Valore'
    )

    # Etichette leggibili
    melted['Tipo'] = melted['Tipo'].map({
        'VALORE': 'Valori Assoluti',
        'VALORE_NORM': 'Valori Normalizzati (base 100)'
    })

    # Plot con LOC_CTP in colonna, tipo in riga
    g = sns.relplot(
        data=melted,
        x='DATA_OSS',
        y='Valore',
        hue='TIPTASSO',
        kind='line',
        col='LOC_CTP',
        row='Tipo',
        facet_kws={'sharey': False},
        height=4,
        aspect=1.5
    )

    # Asse X = solo anno
    for ax in g.axes.flat:
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

    g.set_axis_labels("Anno", "")
    g.set_titles("{row_name} ‚Äî {col_name}")
    if titolo_blocco:
        g.fig.suptitle(titolo_blocco, fontsize=16)
        g.fig.subplots_adjust(top=0.9)  # spazio per il titolo

    plt.tight_layout()
    plt.show()

# --- Esegui i due blocchi senza duplicare codice ---
plot_blocco_nuts(['ITF1', 'ITF2', 'ITF3', 'ITF4'], titolo_blocco="Mezzogiorno - Parte 1")
plot_blocco_nuts(['ITF5', 'ITF6', 'ITG1', 'ITG2'], titolo_blocco="Mezzogiorno - Parte 2")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Definizione gruppi
gruppo1 = pug
gruppo2 = []
tutti = gruppo1 + gruppo2

# Filtro dati
filtered = data[
    (data['LOC_CTP'].isin(tutti)) &
    (data['ENTE_SEGN'] == '1070001') &
    (data['DESINV'].isin([905])) &
    (data['TIPTASSO'].isin([80, 800]))
][['DATA_OSS', 'LOC_CTP', 'TIPTASSO', 'VALORE']].copy()

# Assegna gruppo
filtered['Gruppo'] = filtered['LOC_CTP'].apply(lambda x: 'Gruppo 1' if x in gruppo1 else 'Gruppo 2')

# Converti data
filtered['DATA_OSS'] = pd.to_datetime(filtered['DATA_OSS'])

# Normalizzazione base 100
filtered['VALORE_NORM'] = (
    filtered.sort_values('DATA_OSS')
    .groupby(['LOC_CTP', 'TIPTASSO'])['VALORE']
    .transform(lambda x: (x / x.iloc[0]) * 100)
)

# Reshape per tipo
melted = filtered.melt(
    id_vars=['DATA_OSS', 'LOC_CTP', 'TIPTASSO', 'Gruppo'],
    value_vars=['VALORE', 'VALORE_NORM'],
    var_name='Tipo',
    value_name='Valore'
)

# Etichette leggibili
melted['Tipo'] = melted['Tipo'].map({
    'VALORE': 'Valori Assoluti',
    'VALORE_NORM': 'Valori Normalizzati (base 100)'
})

# Ordine fisso per riga
melted['Tipo'] = pd.Categorical(
    melted['Tipo'],
    categories=['Valori Assoluti', 'Valori Normalizzati (base 100)'],
    ordered=True
)

# Crea griglia 4x2
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(18, 12), sharex=True, sharey=False)

# Mappa posizione righe per i gruppi
posizioni = {
    'Gruppo 1': (0, 1),
    'Gruppo 2': (2, 3)
}

# Colori personalizzati per TIPTASSO
colori_tasso = {
    800: '#007a53',
    80: '#FF0000'
}

# Per legenda globale
linee_legenda = {}

# Plot manuale
for i, loc in enumerate(tutti):
    gruppo = 'Gruppo 1' if loc in gruppo1 else 'Gruppo 2'
    loc_data = melted[melted['LOC_CTP'] == loc]
    for j, tipo in enumerate(loc_data['Tipo'].cat.categories):
        ax_row = posizioni[gruppo][j]
        ax_col = i % 4
        ax = axes[ax_row, ax_col]

        subset = loc_data[loc_data['Tipo'] == tipo]
        for tasso in subset['TIPTASSO'].unique():
            serie = subset[subset['TIPTASSO'] == tasso]
            line, = ax.plot(
                serie['DATA_OSS'],
                serie['Valore'],
                label=f"Tasso {tasso}",
                color=colori_tasso[tasso],
                linewidth=2
            )
            # Solo una volta per legenda
            if tasso not in linee_legenda:
                linee_legenda[tasso] = line

        if ax_row == 0 or ax_row == 2:
            ax.set_title(loc, fontweight='bold')
        if ax_col == 0:
            ax.set_ylabel(tipo)
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
        ax.tick_params(axis='x', rotation=45)
        ax.grid(True)

# Legenda in basso
fig.legend(
    handles=[linee_legenda[80], linee_legenda[800]],
    labels=['Tasso 80', 'Tasso 800'],
    loc='lower center',
    ncol=2,
    fontsize=11,
    frameon=False,
    title="Legenda",
    title_fontsize=12
)

fig.suptitle("Valori Assoluti e Normalizzati - regioni del Mezzogiorno", fontsize=16)
plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import os

def plot_mezzogiorno_consistenze(data, desinv, tabella, gruppo1, gruppo2, output_dir='.'):
    # Unione dei due gruppi
    tutti = gruppo1 + gruppo2

    # Filtro dei dati per tasso 80/800 e il DESINV richiesto
    filtered = data[
        (data['LOC_CTP'].isin(tutti)) &
        (data['ENTE_SEGN'] == '1070001') &
        (data['DESINV'] == desinv) &
        (data['TIPTASSO'].isin([80, 800]))
    ][['DATA_OSS', 'LOC_CTP', 'TIPTASSO', 'VALORE']].copy()

    if filtered.empty:
        print(f"[!] Nessun dato per {tabella}, DESINV={desinv}")
        return

    # Aggiunta colonna 'Gruppo'
    filtered['Gruppo'] = filtered['LOC_CTP'].apply(lambda x: 'Gruppo 1' if x in gruppo1 else 'Gruppo 2')

    # Conversione della data
    filtered['DATA_OSS'] = pd.to_datetime(filtered['DATA_OSS'])

    # Normalizzazione base 100 per LOC_CTP + TIPTASSO
    filtered['VALORE_NORM'] = (
        filtered.sort_values('DATA_OSS')
        .groupby(['LOC_CTP', 'TIPTASSO'])['VALORE']
        .transform(lambda x: (x / x.iloc[0]) * 100 if x.iloc[0] != 0 else 0)
    )

    # Reshape in formato long
    melted = filtered.melt(
        id_vars=['DATA_OSS', 'LOC_CTP', 'TIPTASSO', 'Gruppo'],
        value_vars=['VALORE', 'VALORE_NORM'],
        var_name='Tipo',
        value_name='Valore'
    )

    # Etichette leggibili
    melted['Tipo'] = melted['Tipo'].map({
        'VALORE': 'Valori Assoluti',
        'VALORE_NORM': 'Valori Normalizzati (base 100)'
    })

    # Ordine fisso per righe
    melted['Tipo'] = pd.Categorical(
        melted['Tipo'],
        categories=['Valori Assoluti', 'Valori Normalizzati (base 100)'],
        ordered=True
    )

    # Creazione figure e assi
    fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(18, 12), sharex=True, sharey=False)

    # Mappatura righe: Gruppo 1 in alto, Gruppo 2 in basso
    posizioni = {'Gruppo 1': (0, 1), 'Gruppo 2': (2, 3)}
    colori_tasso = {800: '#007a53', 80: '#FF0000'}
    linee_legenda = {}

    # Disegno dei grafici
    for i, loc in enumerate(tutti):
        gruppo = 'Gruppo 1' if loc in gruppo1 else 'Gruppo 2'
        loc_data = melted[melted['LOC_CTP'] == loc]

        for j, tipo in enumerate(melted['Tipo'].cat.categories):
            ax_row = posizioni[gruppo][j]
            ax_col = i % 4
            ax = axes[ax_row, ax_col]
            subset = loc_data[loc_data['Tipo'] == tipo]

            for tasso in [800, 80]:
                serie = subset[subset['TIPTASSO'] == tasso]
                if not serie.empty:
                    line, = ax.plot(
                        serie['DATA_OSS'],
                        serie['Valore'],
                        label=f'Tasso {tasso}',
                        color=colori_tasso[tasso],
                        linewidth=2
                    )
                    if tasso not in linee_legenda:
                        linee_legenda[tasso] = line

            if ax_row in [0, 2]:
                ax.set_title(loc, fontweight='bold')
            if ax_col == 0:
                ax.set_ylabel(tipo)

            ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
            ax.tick_params(axis='x', rotation=45)
            ax.grid(True)


    # Legenda globale
    fig.legend(
        handles=[linee_legenda[80], linee_legenda[800]],
        labels=['Tasso 80', 'Tasso 800'],
        loc='lower center',
        ncol=2,
        fontsize=11,
        frameon=False,
        title="Legenda",
        title_fontsize=12
    )


    # Titolo e layout
    fig.suptitle(f"{tabella} ‚Äì DESINV {desinv} ‚Äì Valori Assoluti e Normalizzati", fontsize=16)
    plt.tight_layout(rect=[0, 0.05, 1, 0.95])

    # Salvataggio PNG
    os.makedirs(output_dir, exist_ok=True)
    output_path = os.path.join(output_dir, f"{tabella}_DESINV_{desinv}.png")
    plt.savefig(output_path, dpi=300)
    plt.close()
    print(f"[‚úì] Salvato: {output_path}")
```

```python
tabelle = ['TFR10420', 'TFR10430']
desinvs = [997,905]
gruppo1 = ['ITF2','ITF22', 'ITF21']
gruppo2 = []

for tabella in tabelle:
    df = pd.read_sql(f"SELECT * FROM {tabella}", sqlitecon)
    for desinv in desinvs:
        plot_mezzogiorno_consistenze(df, desinv, tabella, gruppo1, gruppo2)
```
