```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtro i dati
impieghi = data[
    (data['Elemento:1'] == 'SBI59') & 
    (data['Elemento'] == '52000700')
][['DATA_OSS', 'Descrizione', 'Descrizione:2', 'VALORE']].copy()

# Converto DATA_OSS in datetime (se necessario)
impieghi['DATA_OSS'] = pd.to_datetime(impieghi['DATA_OSS'])

# Ordino per data crescente
impieghi = impieghi.sort_values(by='DATA_OSS')

# Line plot
plt.figure(figsize=(10, 5))
plt.plot(impieghi['DATA_OSS'], impieghi['VALORE'], marker='o')
plt.title('Evoluzione degli impieghi SBI59  Codice 52000700')
plt.xlabel('Data di Osservazione')
plt.ylabel('Valore')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
